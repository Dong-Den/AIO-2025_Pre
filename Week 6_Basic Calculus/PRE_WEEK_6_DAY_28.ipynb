# **Calculus in Machine Learning - Basic Derivatives**

# phương pháp sai phân hữu hạn

```python
import numpy as np
import matplotlib.pyplot as plt

def numerical_derivative(f,x, h=1e-5):
    return (f(x+h)-f(x))/h

def plot_function_and_derivative(f, x_range=(-2,2), title="Đồ thị f(x) và f'(x)"):
    x_vals = np.linspace(x_range[0], x_range[1], 400)
    y_vals = f(x_vals)
    y_derivative = numerical_derivative(f, x_vals)

    plt.figure(figsize=(10,6))
    plt.plot(x_vals, y_vals, label='f(x)')
    plt.plot(x_vals, y_derivative, label="f'(x) (numerical)",color="red", linestyle="--")
    plt.axhline(0,color='black',linewidth=0.5, linestyle='--')
    plt.axvline(0,color='black', linewidth=0.5,linestyle='--')
    plt.xlabel("x")
    plt.ylabel("f9x) & f'(x)")
    plt.title(title)
    plt.legend()
    plt.grid()
    plt.savefig(f'{title}.png')
    plt.show()
```

```python
# Hàm bậc 2:
def f1(x):
    return x**2-3*x+2
plot_function_and_derivative(f1, x_range=(-2,4), title="Đồ thị hàm bậc 2: x^2-3x+2")

# Hàm bậc 3:

def f2(x):
    return x**3-3*x**2+4*x-2
plot_function_and_derivative(f2, x_range=(-2,4), title="Đồ thị hàm bậc 3: x^3- 3x^2 + 4x- 2")

# Hàm số logarit:
def f3(x):
    return np.exp(x)
plot_function_and_derivative(f3, x_range=(-2,4), title="Đồ thị hàm số mũ: e^x")

# Hàm logarit:
def f4(x):
    return np.log(x)
plot_function_and_derivative(f4, x_range=(0.1,4), title="Đồ thị hàm logarit: ln(x)")

# Hàm sin:
def f5(x):
    return np.sin(x)
plot_function_and_derivative(f5, x_range=(-2*np.pi,2*np.pi), title="Đồ thị hàm sin(x)")

# Hàm cos:
def f6(x):
    return np.cos(x)
plot_function_and_derivative(f6, x_range=(-2*np.pi,2*np.pi), title="Đồ thị hàm cos(x)")

# Hàm tan:
def f7(x):
    return np.tan(x)
plot_function_and_derivative(f7, x_range=(-np.pi/2+0.1,np.pi/2-0.1), title="Đồ thị hàm tan(x)")

# Hàm căn bậc hai:
def f8(x):
    return np.sqrt(x)
plot_function_and_derivative(f8, x_range=(0,4), title="Đồ thị hàm căn bậc hai: sqrt(x)")
```
