# Đạo Hàm của Hàm Kích Hoạt Sigmoid

```python
import numpy as np
import matplotlib.pyplot as plt

def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

x = np.array([-2,-1,0,1,2])

print("Sigmoid:", sigmoid(x))
print("Sigmoid Derivative:", sigmoid_derivative(x))
```

```python
x1=np.linspace(-10,10, 100)
y1=sigmoid(x1)
dy1_dx1=sigmoid_derivative(x1)
plt.plot(x1,y1, label="Sigmoid(x1)")
plt.plot(x1,dy1_dx1, label="Sigmoid_derivative(x1)", linestyle='--') # linestyle ='dashed'
plt.legend()
plt.show()
```

# Đạo Hàm của Hàm Tổn Thất MSE
# Mean Squared Error - MSE

```python
import numpy as np
import matplotlib.pyplot as plt

def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

def mse_derivative(y_true, y_pred):
    return -2 * (y_true - y_pred) / len(y_true)

y_true=np.array([3,5,2,8])
y_pred=np.array([2.5, 4.8, 2.1, 7.5])

print("MSE Derivative:", mse_derivative(y_true, y_pred))
```

# Đạo Hàm của Hàm Kích Hoạt ReLU

```python
import numpy as np
import matplotlib.pyplot as plt

def relu(x):
    return np.maximum(0, x)

def relu_derivative(x):
    return np.where(x > 0, 1, 0)

x=np.array([-2,-1,0,1,2])

print("ReLU:", relu(x))
print("ReLU Derivative:", relu_derivative(x))
```

```python
x3=np.linspace(-10,10, 100)
y3=relu(x3)
dy3_dx3=relu_derivative(x3)
plt.plot(x3,y3, label="ReLU(x3)")
plt.plot(x3,dy3_dx3, label="ReLU_derivative(x3)", linestyle='--') # linestyle ='dashed'
plt.legend()
plt.show()
```

# Đạo Hàm của Hàm Softmax và Cross-Entropy

image.png

#### Hàm Softmax chuyển đổi một vector thành một phân phối xác suất:

#### np.exp(z - np.max(z)): Để tránh hiện tượng tràn số khi tính hàm mũ với các giá trị lớn, ta trừ đi giá trị lớn nhất trong vector đầu vào trước khi tính hàm mũ. Điều này giúp ổn định các phép tính số học

#### exp_z/np.sum(exp_z,axis=0): Sau khi tính hàm mũ cho từng phần tử, ta chia mỗi phần tử cho tổng của tất cả các giá trị này để chuẩn hóa thành xác suất

```python
import numpy as np

def softmax(z):
    exp_z=np.exp(z-np.max(z))
    return exp_z/np.sum(exp_z,axis=0)

def softmax_derivative(z):  # Dùng trong Cross-Entropy
    return softmax(z)*(1-softmax(z))  #Gradient của từng già trị

z=np.array([2.0,1.0, 0.1])
print("Softmax:", softmax(z))
print("Softmax Gradient:", softmax_derivative(z))
```

# Đạo Hàm của Hàm Kích Hoạt Tanh

```python
import numpy as np
import matplotlib.pyplot as plt

def tanh(x):
    return np.tanh(x)
def tanh_derivative(x):
    return 1 - np.tanh(x)**2

x=np.array([-2,-1,0,1,2])
print("Tanh:", tanh(x))
print("Tanh Gradient:", tanh_derivative(x))
```

```python
x5=np.linspace(-4,4,100)
y5=tanh(x5)
dy5_dx5=tanh_derivative(x5)
plt.plot(x5,y5, label="Tanh(x)")
plt.plot(x5,dy5_dx5, label="Tanh_derivative(x)", linestyle="--")
plt.legend()
plt.show()
```
