# Phương pháp sai phân hữu hạn trong hàm số hợp

```python
import numpy as np
import matplotlib.pyplot as plt

def numerical_derivative(f, x, h=1e-5):
    return (f(x + h) - f(x)) / h

def plot_chain_rule(f, g, x_range=(-2,2), title= "Đạo hàm Chain Rule"):
    x_vals = np.linspace(x_range[0], x_range[1], 140)

    g_vals = g(x_vals)
    f_vals=f(g_vals)
    chain_rule_derivative_vals = numerical_derivative(f,g(x_vals)) * numerical_derivative(g, x_vals)

    # Avoid division by zero if g(x+h) - g(x) is too small
    #The original code had an incorrect syntax for the if statement
    #Additionally, it was missing an import for numpy and matplotlib

    plt.figure(figsize=(8, 5))
    plt.plot(x_vals, f_vals, label="f(g(x))", color="blue")
    plt.plot(x_vals,chain_rule_derivative_vals, label="f'(x) theo Chain rule", color='red', linestyle="--")
    plt.axhline(0, color='black', linewidth=0.5, linestyle="--")
    plt.axvline(0, color='black', linewidth=0.5, linestyle="--")
    plt.xlabel("x")
    plt.ylabel("Giá trị")
    plt.title(title)
    plt.legend()
    plt.grid(True)
    plt.show()


```

```python
#Hàm mũ hợp
import numpy as np

def f1(x):
    return np.exp((x**3)-2*x+1)
plot_chain_rule(lambda x:np.exp(x),lambda x:(x**3)-2*x+1, x_range=(-2,2), title= "Đạo hàm hàm mũ hợp ")

# lambda x: Sử dụng để định nghĩa hàm vô danh. Nếu bạn không sử dụng lambda, bạn sẽ phải định nghĩa hàm trước.
# Định nghĩa hàm như bài làm bên dưới
#Lý do cần lambda: Giúp code ngắn gọn hơn, không cần định nghĩa hàm riêng biệt.
# Hữu ích khi chỉ cần sử dụng hàm một lần duy nhất. dùng một lần trong đối số của một hàm khác
# dùng nhiều trong lập trình hàm
# có thể bỏ lambda nếu: đã định nghĩa hàm trước đó/ sử dụng lại hàm nhiều lần trong chương trình
```

```python
#Hàm mũ hợp
import numpy as np

def f1(x):
    return np.exp((x**3)-2*x+1)

def f(x):
    return np.exp(x)

def g(x):
    return (x**3) - 2*x + 1

plot_chain_rule(f, g, x_range=(-2,2), title= "Đạo hàm hàm mũ hợp ")
```
