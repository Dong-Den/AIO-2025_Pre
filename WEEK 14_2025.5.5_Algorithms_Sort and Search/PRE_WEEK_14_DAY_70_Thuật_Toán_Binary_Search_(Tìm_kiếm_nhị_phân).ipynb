```python
# Bài 1: Tìm vị trí phần tử
# Viết hàm binary_search(arr, x) trả về chỉ số của x nếu có, ngược lại trả về -1.

def binary_search(arr, x):
    left = 0
    right = len(arr) - 1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            left = mid + 1
        else:
            right = mid - 1
    return -1

arr = [1, 3, 5, 7, 9]
binary_search(arr, 5)
```

```python
# Bài 2: Tìm phần tử đầu tiên xuất hiện
# Viết hàm tìm chỉ số đầu tiên của phần tử x trong mảng có thể chứa nhiều phần tử giống nhau.

def first_binary_search(arr, x):
    left = 0
    right = len(arr) - 1
    result = -1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] == x:
            result = mid
            right = mid - 1  # tiếp tục tìm bên trái để xác định vị trí đầu tiên cần tìm
        elif arr[mid] < x:
            left = mid + 1
        else:
            right = mid - 1
    return result

arr = [1, 2, 4, 4, 4, 5]
first_binary_search(arr, 4)
```

```python
# Bài 3: Tìm phần tử nhỏ nhất ≥ x
# Trả về chỉ số của phần tử đầu tiên trong mảng lớn hơn hoặc bằng x.

def lower_bound(arr, x):
    left = 0
    right = len(arr) -1
    result = -1
    while left <= right:
        mid = (left + right) // 2
        if arr[mid] >= x:
            result = mid
            right = mid - 1
        else:
            left = mid + 1

    return result

arr = [1, 3, 5, 7, 9]
print(lower_bound(arr, 6))
print(lower_bound(arr, 10))
```

```python
# Bài 4: Đếm số lần gọi đệ quy
# Viết lại Binary Search dưới dạng đệ quy, và trả về số lần gọi hàm tìm kiếm.

call_count = 0

def binary_search_recursive(arr, x, left, right):
    global call_count
    call_count += 1
    if left > right:
        return -1
    mid = (left + right) // 2
    if arr[mid] == x:
        return mid
    elif arr[mid] < x:
        return binary_search_recursive(arr, x, mid + 1, right)
    else:
        return binary_search_recursive(arr, x, left, mid - 1)

arr = [1, 3, 5, 7, 9]

result = binary_search_recursive(arr, 3, 0, len(arr) - 1)
print("số lần gọi hàm tìm kiếm:", call_count)
print("SỐ cần tìm tại vị trí chỉ mục thứ:", result)
```

```python
# Không gọi Đệ quy, chỉ dùng vòng lặp for/while thông thường

call_count0 = 0
def binary_search_iterative(arr, x):
    left = 0
    right = len(arr) - 1
    global call_count0
    while left <= right:
        call_count0 +=1
        mid = (left + right) // 2
        if arr[mid] == x:
            return mid
        elif arr[mid] < x:
            left = mid + 1
        else:
            right = mid - 1
    return -1


arr = [1, 3, 5, 7, 9]

result0 = binary_search_iterative(arr, 3)
print("số lần gọi hàm tìm kiếm:", call_count0)
print("SỐ cần tìm tại vị trí chỉ mục thứ:", result0)
```
