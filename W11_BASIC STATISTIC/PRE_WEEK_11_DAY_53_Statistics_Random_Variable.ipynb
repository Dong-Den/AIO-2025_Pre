# BÀi 1: Sinh Nhãn Phân Loại Ngẫu Nhiên
Sinh 1000 nhãn thuộc 3 lớp cat, dog, rabbit với xác suất tương ứng là 0.4, 0.4, 0.2.

In 10 nhãn đầu và tần suất xuất hiện

```python
import numpy as np
labels = np.random.choice(['cat', 'dog', 'rabbit'], size= 1000, p=[0.4, 0.4, 0.2])
unique, counts = np.unique(labels, return_counts= True)

print("10 nhãn đầu tiên:", labels[:10],"\n")
print("Tần suất xuất hiện của các nhãn:")
for label, count in zip(unique, counts):
    print(f"{label}: {count} . {count/len(labels):.2%}")
```

# Bài 2: Sinh Noise Gaussian
Sinh 100 vector ngẫu nhiên (shape: (100,10)) từ phân phối chuẩn N (0, 1). Tính giá trị trung bình và độ lệch chuẩn của toàn mảng.

```python
import matplotlib.pyplot as plt
import numpy as np

noise = np.random.normal(loc = 0, scale = 1, size= (100,10)) #(batch_size, latent_dim)
mean_noise= np.mean(noise)
std_noise = np.std(noise)

print("Giá trị trung bình của noise:", mean_noise)
print("Độ lệch chuẩn của noise:", std_noise)

plt.figure(figsize=(10,6))
plt.hist(noise.flatten(), bins=30, edgecolor= "black")
plt.title("Histogram of Noise")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```

# Bài 3: Ước Lượng Xác Suất Từ Phân Phối Chuẩn
#### Mô tả: Mô phỏng 10.000 điểm từ phân phối chuẩn N (0, 1) và ước lượng xác suất P (−1 < X < 1)

```python
import numpy as np
x = np.random.normal(loc=0, scale=1, size=10000)
count_in_range = np.sum(np.logical_and(x>-1, x<1))
probability_estimate = count_in_range / len(x)
print("Ước lượng xác suất P(-1< x <1):", probability_estimate)
```

# Bài 4: Mô Phỏng Phân Phối PMF (Probability Mass Function) Hàm phân phối xác suất của biến ngẫu nhiên rời rạc
#### Sinh 1000 nhãn từ các lớp positive, neutral, negative với xác suất tương ứng [0.3, 0.5, 0.2]. Tính PMF thực tế từ dữ liệu sinh ra

```python
import numpy as np
import matplotlib.pyplot as plt
labels = np.random.choice(['positive', 'neutral', 'negative'], size=1000, p=[0.3, 0.5, 0.2])
unique, counts = np.unique(labels, return_counts=True)
pmf= counts / len(labels)
for label, pro in zip(unique, pmf):
    print(f"P({label}): {pro:.2f}")

plt.bar(unique, pmf)
plt.title("Probability Mass Function (PMF) of Labels")
plt.xlabel("Label")
plt.ylabel("Probability")
plt.show()
```

# Bài 5: Sampling Biến Tiềm Ẩn trong VAE
#### Áp dụng công thức reparameterization trick trong VAE
#### z = µ + σ · ϵ, ϵ ∼ N (0, 1)

```python
import numpy as np
mu = np.zeros(10)
sigma = np.ones(10)
eps = np.random.normal(size=10) # LẤy mẫu epsilon từ N(0,1)
z = mu + sigma * eps  # Reparameterization trick
print("Sampled latent vector z:", z)
```

# Biến Ngẫu Nhiên Liên Tục (Continuous Random Variable)

```python
from scipy.stats import norm
import numpy as np
import matplotlib . pyplot as p
x = np.linspace (-4, 4, 1000)
pdf = norm .pdf(x, loc =0, scale =1)
plt.plot (x, pdf , label =" Normal PDF ")
plt.fill_between (x, pdf , where =(x > -1) & (x < 1) , alpha =0.4 , label ="Region [ -1 ,1]")
plt.title (" Continuous Random Variable - Normal Distribution ")
plt.legend ()
plt.grid (True)
plt.show ()
```

# Mô Hình Naive Bayes và Biến Ngẫu Nhiên

```python
from sklearn . naive_bayes import GaussianNB
from sklearn . datasets import make_classification
from sklearn . model_selection import train_test_split
from sklearn . metrics import accuracy_score
X, y = make_classification(n_samples =1000, n_features =10, n_classes =3, n_informative=3, random_state =42)
# Increase n_informative to satisfy the constraint
# n_informative is now 3, making 2**n_informative = 8, which is greater than n_classes * n_clusters_per_class = 6
X_train , X_test , y_train , y_test = train_test_split (X, y, test_size =0.3)
model = GaussianNB ()
model.fit ( X_train , y_train )
y_pred = model . predict ( X_test )
print (" Accuracy :", accuracy_score ( y_test , y_pred ))
```
