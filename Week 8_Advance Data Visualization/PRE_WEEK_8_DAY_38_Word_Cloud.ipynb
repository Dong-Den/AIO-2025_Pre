```python
import pandas as pd
import re
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

data=pd.read_csv('Game_of_Thrones_jon_snow_data.csv', names=['Names'])

def clean_text(Names):
    Names=re.sub(r'^.*?:\s*',' ', Names, flags=re.IGNORECASE)
    Names=Names.replace('\r',' ').replace('\n', ' '). replace('\'', ' ')
    Names=Names.lower()

    return Names

data['cleaned']=data['Names'].astype(str).apply(clean_text)

text_data= ' '.join(data['cleaned'])

print(text_data)

stopwords=set(STOPWORDS)

wordcloud1= WordCloud(background_color = 'white', max_words=200, stopwords=stopwords).generate(text_data)

plt.figure(figsize=(8,10))
plt.imshow(wordcloud1, interpolation='bilinear')
plt.axis('off')

plt.show()
```

```python
import pandas as pd
import re
from wordcloud import WordCloud, STOPWORDS
import matplotlib.pyplot as plt

df = pd.read_csv('Game_of_Thrones_jon_snow_data.csv', names=['Name'])
df['Name'] = df['Name'].replace(r'\n', '', regex=True).replace(r'\r', '', regex=True)
df['Name'] = df['Name'].str.split(':', n=1).str[-1].str.strip()
df['Name'] = df['Name'].replace(r"\' ", '', regex=True)
text = ' '.join(df['Name'])
stopwords = set(STOPWORDS)
wordcloud = WordCloud(width=800, height=400, background_color='white', stopwords=stopwords).generate(text)

plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```

```python
from wordcloud import WordCloud, STOPWORDS
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('Game_of_Thrones_jon_snow_data.csv', names=['Names'])
data['Names'] = data['Names'].str.extract(':(.*?)(?=\r\n)')
text_data = ''.join(data['Names'])
text_data = text_data.replace("\'", "")
stopwords = set(STOPWORDS)
wc = WordCloud(background_color='white', max_words=200, stopwords=stopwords)

wc.generate(text_data)
plt.figure(figsize=(8,10))
plt.imshow(wc, interpolation='bilinear')
plt.axis('off')
plt.show()

import numpy as np
from PIL import Image
img_mask = np.array(Image.open('jon-snow.jpg'))
plt.imshow(img_mask)
plt.axis('off')
plt.show()

wc_mask = WordCloud(background_color='white', mask=img_mask, max_words=200, stopwords=stopwords)
wc_mask.generate(text_data)

plt.figure(figsize=(8,10))
plt.imshow(wc_mask, interpolation='bilinear')
plt.axis('off')
plt.show()
```
