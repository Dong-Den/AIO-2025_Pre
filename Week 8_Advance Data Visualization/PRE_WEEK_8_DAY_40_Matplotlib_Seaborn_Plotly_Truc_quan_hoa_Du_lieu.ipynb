```python
sns.boxplot(data=data, y='Sodium')
plt.title('Biểu đồ BoxPlot kiểm tra Outlier của Sodium')
plt.figure(figsize=(3,2))
plt.show()
```

#### Dữ liệu Sodium có outlier ( upper bound)

#### Loại bỏ tất cả các dòng data có Outlier trong cột Sodium

```python
import numpy as np
import pandas as pd

def remove_outliers_fr(data, sodium):
    Q1 = data[sodium].quantile(0.25)
    Q3 = data[sodium].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    df_no_outliers = data[(data[sodium] >= lower_bound) & (data[sodium] <= upper_bound)]
    return df_no_outliers

df_no_outliers = remove_outliers_fr(data, 'Sodium')
df_no_outliers
```

#### Nhóm dữ liệu Category và đếm theo Item, cho biết mỗi nhóm có bao nhiêu

```python
category_item_counts = data.groupby('Category')['Item'].count()
category_item_counts
```

```python
group_category_item =  data.groupby(by='Category')['Item'].count().to_frame().reset_index()
group_category_item.rename(columns = {'Item':'Item_counts'}, inplace = True)
group_category_item
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(12, 6))
sns.barplot(x='Category', y='Item_counts', data=group_category_item)
# sns.barplot(data=group_category_item, x='Category', y='Item_counts', palette='tab10', hue='Category')
# Dùng bảng màu có 10 màu khác nhau. Màu sắc sẽ phân biệt theo cột Category
plt.title('Số lượng Item có trong Category')
plt.xlabel('Category')
plt.ylabel('Item counts')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
# Xoay nhãn trục X một góc 45 độ để tránh chồng chéo.
# Căn lề phải (horizontal alignment = "right"), giúp nhãn trục dễ đọc hơn
plt.tight_layout() # Adjust layout to prevent labels from overlapping
# tự động điều chỉnh khoảng cách giữa các thành phần trong biểu đồ để tránh tình trạng chồng chéo hoặc bị cắt mất nội dung
plt.show()
```
