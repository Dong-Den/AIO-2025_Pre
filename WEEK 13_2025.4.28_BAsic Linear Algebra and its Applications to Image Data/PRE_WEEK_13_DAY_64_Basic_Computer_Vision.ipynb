```python
# Bi 3: Chuyn nh t BGR sang RGB
# Vit chng trnh chuyn nh t khng gian BGR sang RGB

import cv2
import matplotlib.pyplot as plt
img = cv2.imread('nature.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title('RGB Image')
plt.axis('off')
plt.show()
```

```python
# Bi 4: Chuyn nh t BGR sang HSV
# Chuyn nh gc t BGR sang HSV v hin th kt qu

import cv2
import matplotlib.pyplot as plt
img = cv2.imread('nature.png')
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
plt.title('HSV Image')
plt.axis('off')
plt.show()
```

```python
# Bi 5: Chuyn nh sang Grayscale
# Chuyn nh gc sang nh en trng (grayscale) v lu li vi tn gray.jpg

import cv2
import matplotlib.pyplot as plt
img = cv2.imread('nature.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite('gray.jpg', img_gray)
plt.imshow(img_gray, cmap='gray')
# plt.imshow(cv2.imread('gray.jpg'))
plt.title('Gray Image')
plt.axis('off')
plt.show()
```

```python
# Bi 6: So snh nh RGB v HSV
# Hin th nh gc (RGB) v nh sau khi chuyn HSV cnh nhau. Ghi nhn s khc bit.
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('Image_6385.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# Create a grid plot to display the images
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.ravel()

# Display the original RGB image
axes[0].imshow(img_rgb)
axes[0].set_title('RGB')
axes[0].axis('off')

# Display the HSV image
axes[2].imshow(img_hsv)
axes[2].set_title('HSV')
axes[2].axis('off')

axes[5].axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('nature.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
combined = cv2.hconcat([img_rgb, img_hsv])
plt.imshow(combined)

plt.title('Comparison: RGB (Left) vs. HSV (Right)')
plt.axis('off')
# plt.tight_layout() khng cn thit v  sd hm
plt.show()

#print("Shape ca combined:", combined.shape)
#print("Kiu d liu ca combined:", combined.dtype)

#Gim chiu v ni suy nh
#resized_combined = cv2.resize(combined, (500, 1200), interpolation= cv2.INTER_LINEAR)

#print("Shape ca combined:", resized_combined.shape)
#print("Kiu d liu ca combined:", resized_combined.dtype)
```
