```python
# Bài 1: Đọc và hiển thị ảnh
# Viết chương trình đọc file ảnh tên input.jpg và hiển thị ảnh ra màn hình

import cv2
import matplotlib.pyplot as plt
img = cv2.imread('image.jpg') #Color = BGR
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Bài 2: Lưu ảnh dưới tên khác
# Viết chương trình lưu ảnh vừa đọc thành output.jpg.
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('nature.jpg')

cv2.imwrite('new_nature.jpg', img)
img_new = cv2.imread('new_nature.jpg')
plt.imshow(img_new)
plt.title('New Image')
plt.axis('off')
plt.show()
```

```python
# Bài 3: Chuyển ảnh từ BGR sang RGB
# Viết chương trình chuyển ảnh từ không gian BGR sang RGB

import cv2
import matplotlib.pyplot as plt
img = cv2.imread('nature.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.title('RGB Image')
plt.axis('off')
plt.show()
```

```python
# Bài 4: Chuyển ảnh từ BGR sang HSV
# Chuyển ảnh gốc từ BGR sang HSV và hiển thị kết quả

import cv2
import matplotlib.pyplot as plt
img = cv2.imread('nature.png')
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
plt.imshow(img_hsv)
plt.title('HSV Image')
plt.axis('off')
plt.show()
```

```python
# Bài 5: Chuyển ảnh sang Grayscale
# Chuyển ảnh gốc sang ảnh đen trắng (grayscale) và lưu lại với tên gray.jpg

import cv2
import matplotlib.pyplot as plt
img = cv2.imread('nature.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imwrite('gray.jpg', img_gray)
plt.imshow(img_gray, cmap='gray')
# plt.imshow(cv2.imread('gray.jpg'))
plt.title('Gray Image')
plt.axis('off')
plt.show()
```

```python
# Bài 6: So sánh ảnh RGB và HSV
# Hiển thị ảnh gốc (RGB) và ảnh sau khi chuyển HSV cạnh nhau. Ghi nhận sự khác biệt.
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('Image_6385.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
# Create a grid plot to display the images
fig, axes = plt.subplots(2, 3, figsize=(12, 8))
axes = axes.ravel()

# Display the original RGB image
axes[0].imshow(img_rgb)
axes[0].set_title('RGB')
axes[0].axis('off')

# Display the HSV image
axes[2].imshow(img_hsv)
axes[2].set_title('HSV')
axes[2].axis('off')

axes[5].axis('off')

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('nature.png')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
img_hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
combined = cv2.hconcat([img_rgb, img_hsv])
plt.imshow(combined)

plt.title('Comparison: RGB (Left) vs. HSV (Right)')
plt.axis('off')
# plt.tight_layout() không cần thiết vì đã sd hàm
plt.show()

#print("Shape của combined:", combined.shape)
#print("Kiểu dữ liệu của combined:", combined.dtype)

#Giảm chiều và nội suy ảnh
#resized_combined = cv2.resize(combined, (500, 1200), interpolation= cv2.INTER_LINEAR)

#print("Shape của combined:", resized_combined.shape)
#print("Kiểu dữ liệu của combined:", resized_combined.dtype)
```
