```python
import numpy as np
import matplotlib.pyplot as plt
X= np.array([
    [92.5, 2.1, 65.3],
    [93.2, 2.5, 67.2],
    [91.8, 2.3, 64.0],
    [94.0, 2.8, 70.1],
    [95.2, 3.0, 72.5],
    [96.1, 3.2, 74.3],
    [90.5, 1.8, 61.0],
    [92.0, 2.0, 63.2],
    [89.5, 1.5, 59.8],
    [97.0, 3.5, 76.2],
    [95.8, 3.1, 73.8],
    [94.5, 2.9, 71.5],
    [91.2, 2.2, 62.8],
    [90.0, 1.7, 60.5],
    [98.0, 3.7, 78.0],
    [99.2, 4.0, 80.5],
    [88.5, 1.3, 58.0],
    [87.8, 1.1, 56.5],
    [86.5, 1.0, 55.0],
    [100.0, 4.2, 82.0]
])

y= np.array([
    1800, 1825, 1795, 1850, 1880, 1905, 1750, 1780, 1725, 1925,
    1890, 1860, 1775, 1740, 1950, 1980, 1700, 1680, 1650, 2000
])
class LinearRegression:
    def __init__(self):
        self.w=np.zeros(X.shape[1])
        self.b=0
    def compute_gradients(self,X,y):
        m=X.shape[0] # số lượng mẫu = len(y)
        y_pred=np.dot(X,self.w)+self.b
        dw=(1/m)*np.dot(X.T,(y_pred-y))
        db=(1/m)*np.sum(y_pred-y)
        return dw,db
    def fit(self, X, y, learning_rate=0.001, epochs=10):
        for epoch in range(epochs):
            dw, db = self.compute_gradients(X, y)
            self.w -= learning_rate * dw
            self.b -= learning_rate * db
            y_pred = np.dot(X, self.w) + self.b
            loss = 1/2* np.mean((y_pred - y) ** 2)
            print(f"Epoch {epoch+1}/{epochs}, Loss: {loss:.2f}")
    def predict(self, X):
        return np.dot(X, self.w) + self.b

X_normalized=(X-np.mean(X,axis=0))/np.std(X,axis=0)
model=LinearRegression()
model.fit(X_normalized,y, learning_rate=0.05, epochs=1000)

print("\n Final wieght:",model.w)
print("\n Final bias:",model.b)

y_pred=model.predict(X_normalized)
print("\n So sánh kết quả thực tế và dự đoán:")
for i in range(5):
    print(f"y_true={y[i]}, y_pred={y_pred[i]}")
```

```python
import numpy as np

class LinearRegression:
    def __init__(self, n_features):
        self.w = np.zeros(n_features + 1)  # Thêm 1 cho bias

    def compute_gradients(self, X, y):
        m = X.shape[0]
        X_bias = np.c_[np.ones((m, 1)), X]  # Thêm cột 1 vào X
        y_pred = np.dot(X_bias, self.w)
        dw = (1/m) * np.dot(X_bias.T, (y_pred - y))
        return dw

# Tạo dữ liệu giả lập
X = np.array([[2, 3], [4, 5], [6, 7]])  # 3 mẫu, 2 feature
y = np.array([10, 20, 30])  # Giá trị thực tế

# Khởi tạo mô hình
model = LinearRegression(n_features=2)
gradients = model.compute_gradients(X, y)

print("Gradient:", gradients)
```
